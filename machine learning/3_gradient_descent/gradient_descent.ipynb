```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
def gradient_descent(x,y,iterations = 1000,learning_rate = 0.05):
    m_curr = b_curr = 0
    n = len(x)
    plt.scatter(x,y,color='red',marker='+',linewidth=5)
    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)])
        plt.plot(x,y_predicted,color='green')
        md = -(2/n)*sum(x*(y-y_predicted))
        bd = -(2/n)*sum(y-y_predicted)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        if i%50==1:
            print ("m {}, b {}, cost {}, iteration {}".format(m_curr,b_curr,cost, i))
            
        
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
gradient_descent(x,y)
```

```python
gradient_descent(x,y,learning_rate = 0.1)#here learning rate is too high
```

```python
gradient_descent(x,y,learning_rate = 0.08)
```

```python
gradient_descent(x,y,iterations = 1000,learning_rate = 0.07)
```
